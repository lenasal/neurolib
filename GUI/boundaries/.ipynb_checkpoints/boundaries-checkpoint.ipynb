```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pickle

from neurolib.models.aln import ALNModel
from neurolib.utils import plotFunctions as plotFunc
from neurolib.utils import costFunctions as cost
import neurolib.dashboard.functions as functions
import neurolib.dashboard.data as data
    
# This will reload all imports as soon as the code changes
%load_ext autoreload
%autoreload 2 

#path = os.path.join(os.getcwd(), "plots")
```

### Bistability

```python
aln = ALNModel()
data.set_parameters(aln)

state_vars = aln.state_vars
init_vars = aln.init_vars

##############################################################
def setinit(init_vars_):
    for iv in range(len(init_vars)):
        for sv in range(len(state_vars)):
            if state_vars[sv] in init_vars[iv]:
                #print("set init vars ", )
                if aln.params[init_vars[iv]].ndim == 2:
                    aln.params[init_vars[iv]][0,:] = init_vars_[sv]
                else:
                    aln.params[init_vars[iv]][0] = init_vars_[sv]
```

```python
lower_bound_bi_e = []
lower_bound_bi_i = []
upper_bound_bi_e = []
upper_bound_bi_i = []
bi_e = []
bi_i = []
```

```python
range_e = np.arange(0.3, 0.81, 0.025)
range_i = np.arange(0.275, 0.901, 0.025)

aln.params['duration'] = 6000.
control0 = aln.getZeroControl()
control0 = functions.step_control(aln, maxI_ = 3.)

for i in range_i:
    print('---------', i)
    lower_bound = False
    upper_bound = False
    bistable = False
    for e in range_e:
        #print('---------', e)
        bistable = False
        
        aln.params.mue_ext_mean = e * 5.
        aln.params.mui_ext_mean = i * 5.
        
        aln.run(control=control0)
        
        mean1_1 = np.mean(aln.rates_exc[0,27000:28000])
        mean1_2 = np.mean(aln.rates_exc[0,28000:29000])
        mean1_3 = np.mean(aln.rates_exc[0,29000:30000])
        if np.abs(mean1_1 - mean1_2) < 1e-4 and np.abs(mean1_1 - mean1_3) < 1e-4:
            stable1 = True
        
        mean2_1 = np.mean(aln.rates_exc[0,57000:58000])
        mean2_2 = np.mean(aln.rates_exc[0,58000:59000])
        mean2_3 = np.mean(aln.rates_exc[0,59000:60000])
        
        if np.abs(mean2_1 - mean2_2) < 1e-4 and np.abs(mean2_1 - mean2_3) < 1e-4:
            stable2 = True
            
        if stable1 and stable2 and np.abs(mean1_1 - mean2_1) > 1.:
            bistable = True
            
        #print(bistable, lower_bound, upper_bound)
            
        if bistable:
            #plotFunc.plot_traces(aln, control0)
            upper_bound = True
            bi_e.append(e)
            bi_i.append(i)
            if np.abs(bi_e[-1] - 0.4) < 1e-4 and np.abs(bi_i[-1] - 0.325) < 1e-4:
                    bi_e.pop()
                    bi_i.pop()
            
            if lower_bound == False:
                lower_bound_bi_e.append(e)
                lower_bound_bi_i.append(i)
                lower_bound = True
                if np.abs(lower_bound_bi_e[-1] - 0.4) < 1e-4 and np.abs(lower_bound_bi_i[-1] - 0.325) < 1e-4:
                    lower_bound_bi_e.pop()
                    lower_bound_bi_i.pop()
                
        if not bistable:
            if upper_bound:
                upper_bound = False
                upper_bound_bi_e.append(e-0.025)
                upper_bound_bi_i.append(i)
                if np.abs(upper_bound_bi_e[-1] - 0.4) < 1e-4 and np.abs(upper_bound_bi_i[-1] - 0.325) < 1e-4:
                    upper_bound_bi_e.pop()
                    upper_bound_bi_i.pop()
                
            
plt.scatter(bi_e, bi_i, s=60, marker='x', c="grey")
plt.scatter(lower_bound_bi_e, lower_bound_bi_i, s=12, c="red")
plt.scatter(upper_bound_bi_e, upper_bound_bi_i, s=12, c="blue")

val = []
for i,j in zip(bi_e, bi_i):
    if [i,j] in val:
        print("already there", i,j)
    val.append([i,j])
```

```python
boundary_bi_e = []
boundary_bi_i = []

for index in range(len(upper_bound_bi_e)-1, -1, -1):
    boundary_bi_e.append(upper_bound_bi_e[index])
    boundary_bi_i.append(upper_bound_bi_i[index])
for index in range(len(lower_bound_bi_e)):
    boundary_bi_e.append(lower_bound_bi_e[index])
    boundary_bi_i.append(lower_bound_bi_i[index])
    
plt.scatter(boundary_bi_e, boundary_bi_i, s=12, c="blue")
```

```python
with open('boundary_bi.pickle','wb') as f:
    pickle.dump([boundary_bi_e, boundary_bi_i], f)
    
with open('bi.pickle','wb') as f:
    pickle.dump([bi_e, bi_i], f)
```

```python
lower_bound_LC_e = []
lower_bound_LC_i = []
upper_bound_LC_e = []
upper_bound_LC_i = []
LC_e = []
LC_i = []
```

```python
range_e = np.arange(0.0, 0.51, 0.025)
range_i = np.arange(0.0, 0.301, 0.025)

aln.params['duration'] = 6000.
control0 = aln.getZeroControl()
control0 = functions.step_control(aln, maxI_ = 3.)

for i in range_i:
    print('---------', i)
    lower_bound = False
    upper_bound = False
    for e in range_e:
        #print('---------', e)
        stable1 = False
        stable2 = False
        stable3 = False
        
        aln.params.mue_ext_mean = e * 5.
        aln.params.mui_ext_mean = i * 5.
        
        aln.run(control=control0)
        
        mean1_1 = np.mean(aln.rates_exc[0,27050:28000])
        mean1_2 = np.mean(aln.rates_exc[0,28050:29000])
        mean1_3 = np.mean(aln.rates_exc[0,29050:30000])
        #print(mean1_1, mean1_2, mean1_3)
        if np.abs(mean1_1 - mean1_2) < 1e-4 and np.abs(mean1_1 - mean1_3) < 1e-4:
            stable1 = True
        
        mean2_1 = np.mean(aln.rates_exc[0,57050:58000])
        mean2_2 = np.mean(aln.rates_exc[0,58050:59000])
        mean2_3 = np.mean(aln.rates_exc[0,59050:60000])
                
        if np.abs(mean2_1 - mean2_2) < 1e-4 and np.abs(mean2_1 - mean2_3) < 1e-4:
            stable2 = True
            
        if stable1 and stable2 and np.abs(mean1_1 - mean2_1) < 1e-4:
            stable3 = True
            
        #print(stable1, stable2, stable3)
            
        if not stable3:
            #plotFunc.plot_traces(aln, control0)
            upper_bound = True
            LC_e.append(e)
            LC_i.append(i)
            
            if lower_bound == False:
                lower_bound_LC_e.append(e)
                lower_bound_LC_i.append(i)
                lower_bound = True
                
        if stable3:
            if upper_bound:
                upper_bound = False
                upper_bound_LC_e.append(e-0.025)
                upper_bound_LC_i.append(i)
                
        #print(bi_e, lower_bound_bi_e, upper_bound_bi_e)
                
            
plt.scatter(LC_e, LC_i, s=60, marker='x', c="grey")
plt.scatter(lower_bound_LC_e, lower_bound_LC_i, s=12, c="red")
plt.scatter(upper_bound_LC_e, upper_bound_LC_i, s=12, c="blue")

# 0.325, 0.275 is osc
# 0.35, 0.275 is osc
# 0.35, 0.3 is down state
# 0.375, 0.3 is osc
# 0.35, 0.325 is down state
# 0.375, 0.325 is down state
# 0.4, 0.325 is bistable LC, up
# 0.425, 0.325 is up
```

```python
boundary_LC_e = []
boundary_LC_i = []

for index in range(len(lower_bound_LC_e)):
    boundary_LC_e.append(lower_bound_LC_e[index])
    boundary_LC_i.append(lower_bound_LC_i[index])
for index in range(len(upper_bound_LC_e)-1, -1, -1):
    boundary_LC_e.append(upper_bound_LC_e[index])
    boundary_LC_i.append(upper_bound_LC_i[index])

    
plt.scatter(boundary_LC_e, boundary_LC_i, s=12, c="blue")
```
