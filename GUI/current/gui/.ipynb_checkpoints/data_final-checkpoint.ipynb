```python
import os
import pickle
import numpy as np
```

```python
### comprise data from three files:

# TASK 1

cases_ = ['00101', '00111', '00121']
final_file_ = '.' + os.sep + 'data_final' + os.sep + 'control_1.pickle'

file_ = '.' + os.sep + 'data' + os.sep + cases_[0] + os.sep + 'control_000' + cases_[0][-2:] + '.pickle'
final_file_init = '.' + os.sep + 'data_final' + os.sep + 'control_1_init.pickle'

if os.path.isfile(file_) :
    print("file found")
    
    with open(file_,'rb') as f:
        load_array = pickle.load(f)

    bestControl_0 = load_array[0]
    bestState_0 = load_array[1]
    cost_0 = load_array[2]
    runtime_0 = load_array[3]
    grad_0 = load_array[4]
    phi_0 = load_array[5]
    costnode_0 = load_array[6]
    weights_0 = load_array[7]
    
file_ = '.' + os.sep + 'data' + os.sep + cases_[1] + os.sep + 'control_000' + cases_[1][-2:] + '.pickle'

if os.path.isfile(file_) :
    print("file found")
    
    with open(file_,'rb') as f:
        load_array = pickle.load(f)

    bestControl_1 = load_array[0]
    bestState_1 = load_array[1]
    cost_1 = load_array[2]
    runtime_1 = load_array[3]
    grad_1 = load_array[4]
    phi_1 = load_array[5]
    costnode_1 = load_array[6]
    weights_1 = load_array[7]
    
file_ = '.' + os.sep + 'data' + os.sep + cases_[2] + os.sep + 'control_000' + cases_[2][-2:] + '.pickle'

if os.path.isfile(file_) :
    print("file found")
    
    with open(file_,'rb') as f:
        load_array = pickle.load(f)

    bestControl_2 = load_array[0]
    bestState_2 = load_array[1]
    cost_2 = load_array[2]
    runtime_2 = load_array[3]
    grad_2 = load_array[4]
    phi_2 = load_array[5]
    costnode_2 = load_array[6]
    weights_2 = load_array[7]
    
bestControl_init = [None] * len(bestControl_0)
bestState_init = [None] * len(bestControl_0)
cost_init = [None] * len(bestControl_0)
runtime_init = [None] * len(bestControl_0)
grad_init = [None] * len(bestControl_0)
phi_init = [None] * len(bestControl_0)
costnode_init = [None] * len(bestControl_0)
weights_init = [None] * len(bestControl_0)
    
bestControl = [None] * len(bestControl_0)
bestState = [None] * len(bestControl_0)
cost = [None] * len(bestControl_0)
runtime = [None] * len(bestControl_0)
grad = [None] * len(bestControl_0)
phi = [None] * len(bestControl_0)
costnode = [None] * len(bestControl_0)
weights = [None] * len(bestControl_0)
    
for i in range(len(bestControl_0)):
    if type(bestControl_0[i]) == type(None):
        continue
        
    print(i)
    j = -1
    while cost_0[i][-j] == 0.:
        j -= 1
    cost0 = cost_0[i][-j]
    
    j = -1
    while cost_1[i][-j] == 0.:
        j -= 1
    cost1 = cost_1[i][-j]
    
    j = -1
    while cost_2[i][-j] == 0.:
        j -= 1
    cost2 = cost_2[i][-j]
    
    if np.amin([cost0, cost1, cost2]) == cost0:
        bestControl[i] = bestControl_0
        bestState[i] = bestState_0
        cost[i] = cost_0
        runtime[i] = runtime_0
        grad[i] = grad_0
        phi[i] = phi_0
        costnode[i] = costnode_0
        weights[i] = weights_0
        
        file_init = '.' + os.sep + 'data' + os.sep + cases_[0] + os.sep + 'control_init_001' + cases_[0][-2:] + '.pickle'
    
    elif np.amin([cost0, cost1, cost2]) == cost1:
        bestControl[i] = bestControl_1
        bestState[i] = bestState_1
        cost[i] = cost_1
        runtime[i] = runtime_1
        grad[i] = grad_1
        phi[i] = phi_1
        costnode[i] = costnode_1
        weights[i] = weights_1
        
        file_init = '.' + os.sep + 'data' + os.sep + cases_[1] + os.sep + 'control_init_001' + cases_[1][-2:] + '.pickle'
        
    elif np.amin([cost0, cost1, cost2]) == cost2:
        bestControl[i] = bestControl_2
        bestState[i] = bestState_2
        cost[i] = cost_2
        runtime[i] = runtime_2
        grad[i] = grad_2
        phi[i] = phi_2
        costnode[i] = costnode_2
        weights[i] = weights_2
        
        file_init = '.' + os.sep + 'data' + os.sep + cases_[2] + os.sep + 'control_init_001' + cases_[2][-2:] + '.pickle'
    
    if os.path.isfile(file_init) :
        print("init file found")
    
    with open(file_init,'rb') as f:
        load_array = pickle.load(f)

    bestControl_init[i] = load_array[0]
    bestState_init[i] = load_array[1]
    cost_init[i] = load_array[2]
    runtime_init[i] = load_array[3]
    grad_init[i] = load_array[4]
    phi_init[i] = load_array[5]
    costnode_init[i] = load_array[6]
    weights_init[i] = load_array[7]
    
with open(final_file_,'wb') as f:
    pickle.dump([bestControl, bestState, cost, runtime, grad, phi, costnode, weights], f)
with open(final_file_init,'wb') as f:
    pickle.dump([bestControl_init, bestState_init, cost_init, runtime_init, grad_init, phi_init, costnode_init, weights_init], f)
```
