```python
import numpy as np
import matplotlib.pyplot as plt
from neurolib.utils import costFunctions as cost
    
# This will reload all imports as soon as the code changes
%load_ext autoreload
%autoreload 2 

cost.setParams(0., 0., 1.)
```

```python
def plot_control(control_, T, dt, N, var):
    
    fig, ax = plt.subplots(N, var, figsize=(6, 6))
    for n in range(N):
        for v in range(var):
            ax[n,v].plot(np.arange(0, int(T), dt), control_[n,v,:], label = "node = " + str(n) + ", var = " + str(v) )
            ax[n,v].legend()
    #plt.plot(np.arange(0, int(T), dt), cost.f_cost_sparsity1(control_), label = "sum of absolute control input")
    #plt.legend()
    plt.show()
```

```python
T = 10.
dt = 0.1
N = 2
var = 2
```

```python
control1 = np.zeros(( N, var, int(T/dt) ))  
for t in range(control1.shape[2]):
    control1[:,:,t] = t / 25.
    
#print(1.6 * np.sqrt(1000./3.))

print("cost with functional j1 = ", cost.f_cost_sparsity1_int(dt, control1))
print("cost with functional j2 = ", cost.f_cost_sparsity2_int(dt, control1))
print("cost with functional j3 = ", cost.f_cost_sparsity3_int(dt, control1))
print("cost with functional j4 = ", cost.f_cost_sparsity4_int(dt, control1))
plot_control(control1, T, dt, N, var)
```
